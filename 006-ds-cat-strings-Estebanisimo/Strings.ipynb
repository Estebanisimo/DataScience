### Escape sequences

"`\n`" és l'anomenada "seqüència d'escapament". Una seqüència d'escapada és un caràcter de cadena escrit com una barra invertida seguit d'un codi, que pot ser un o diversos caràcters. Python interpreta les seqüències d'escapada en una cadena com un caràcter especial; un caràcter de control.

```python
word1 = "orange"
word2 = "banana"

def add_newline_between_words(word1,word2):
    new_line = word1 + "\n" + word2
    return(new_line)
    
print(add_newline_between_words(word1,word2))
```

A més del caràcter de nova línia, hi ha més caràcters especials "`\'`" i "`\"`", que es poden utilitzar per col·locar una cometa simple o doble en una cadena, independentment dels caràcters que envolten la cadena. També he esmentat que podeu utilitzar "`\\`" per inserir una barra invertida "real" en una cadena.

Hi ha algunes "seqüències de barra invertida" més que porten a un caràcter especial. La majoria d'aquests són arcaics i no cal que us preocupeu per ells. Els que vull esmentar són "`\t`" que representa una sola tabulació (també coneguda com a "tab").

```python
print("hello \t world") #\t hace de tabulador
```

```python
def place_word_between_single_quotes(w1):
    new_line = '\'' + w1 + "\'"
    return(new_line)

print(place_word_between_single_quotes("miramira"))


def place_word_between_double_quotes(w1):
    new_line = '\t' + w1 + '"'
    return(new_line)

print(place_word_between_double_quotes("dididi"))
```

Informació addicional per als estudiants que vulguin saber-ne més, però no necessària per a aquest curs:

Hi ha un altre caràcter "`\xnn`" pel qual `nn` significa dos dígits hexadecimals, que representa el caràcter amb el número hexadecimal `nn`. Per exemple, "`\x20`" és el caràcter expressat pel número hexadecimal `20`, que és el mateix que el número decimal `32`, que és l'espai (això s'explicarà més endavant en aquest capítol).

En cas que no hàgiu après mai sobre el recompte hexadecimal: els hexadecimals utilitzen un esquema de numeració que utilitza 16 dígits diferents. Utilitzem deu (de "0" a "9"), el binari en fa servir dos (de "0" a "1"), i hexidecimal fa servir de "0" a "9" i després continua de "A" a "F". Una traducció directa d'hexadecimals a decimals converteix "A" en "10", "B" en "11", etcètera. En el recompte decimal, el valor d'un nombre de diversos dígits es troba multiplicant els dígits augmentant les potències de "10", de dreta a esquerra, per exemple, el nombre "1426" és "6 + 2*10 + 4*100 + 1*1000`. Per als nombres hexadecimals, feu el mateix, però multipliqueu per potències de "16", per exemple, el nombre hexadecimal "4AF2" és "2 + 15*16 + 10*256 + 4*4096". Als programadors acostumen a agradar els nombres hexadecimals, ja que els ordinadors treballen amb bytes com a unitat més petita d'emmagatzematge de memòria, i un byte pot emmagatzemar 256 valors diferents, és a dir, qualsevol valor de byte es pot expressar amb un nombre hexadecimal de dos dígits.

### Accés als caràcters d'una cadena

Com he mostrat diverses vegades abans, una cadena és una col·lecció de caràcters en un ordre específic. Podeu accedir als caràcters individuals d'una cadena mitjançant índexs.

### String indices

Cada símbol d'una cadena té una posició, aquesta posició es pot referir al número d'índex de la posició. Els números d'índex comencen a 0 i després augmenten fins a la longitud de la cadena. La taula següent mostra la paraula "taronja" a la primera fila i els índexs de cada lletra a la segona i tercera fila:

&nbsp;&nbsp;__`  o  r  a  n  g  e`__<br>
&nbsp;&nbsp;`  0  1  2  3  4  5`<br>
` -6 -5 -4 -3 -2 -1`

Com podeu veure, podeu utilitzar índexs positius, que comencen a la primera lletra de la cadena i augmenten fins arribar al final de la cadena, o índexs negatius, que comencen amb -1 per a l'última lletra de la cadena i disminueixen fins a s'arriba a la primera lletra de la cadena.

Com que la longitud d'una cadena `s`és `len(s)`, l'última lletra de la cadena té l'índex `len(s)-1`. Amb índexs negatius, la primera lletra de la cadena té l'índex `-len(s)`.

Si s'emmagatzema una cadena en una variable, es pot accedir a les lletres individuals de la cadena mitjançant el nom de la variable i l'índex de la lletra sol·licitada entre claudàtors (`[]`) al costat.

```python
fruit = "orange"

def print_indices(fruit,n):
    print(fruit[n])
    
print_indices(fruit,1) 
print_indices(fruit,2) 
print_indices(fruit,4)
print_indices(fruit,-1)
print_indices(fruit,-6)
print_indices(fruit,-3)
```

A més d'utilitzar índexs únics, també podeu accedir a una subcadena (també anomenada "slice") des d'una cadena utilitzant dos números entre claudàtors amb dos punts (`:`) entremig. El primer d'aquests nombres és l'índex on comença la subcadena, el segon on acaba. La subcadena *no* inclou la lletra del segon índex. En deixar de banda el número esquerre, indiqueu que la subcadena comença al principi de la cadena (és a dir, a l'índex 0). En deixar de banda el número correcte, indiqueu que la subcadena va fins i inclou l'últim caràcter de la cadena.

Si intenteu accedir a un caràcter mitjançant un índex que està fora de l'abast d'una cadena, obtindreu un error d'execució ("índex fora dels límits"). Per a una sèrie d'índexs per accedir a subcadenes, aquestes limitacions no existeixen; podeu utilitzar números que estiguin fora dels límits de la cadena.

```python
fruit = "orange"
print(fruit[:])
print(fruit[0:])
print(fruit[:5])
print(fruit[:100])
print(fruit[:len(fruit)])
print(fruit[1:-1])
print(fruit[2], fruit[1:6])
```

### Traversing strings

Ja hem vist com podeu recórrer els caràcters d'una cadena utilitzant un bucle `for`:

```python
fruit = 'apple'

def traverse_characters(word):
    new_word = ""
    for char in word:
        new_word+=(char + ' - ')
    return new_word
print(traverse_characters(fruit))
```

Ara que coneixeu els índexs, probablement us adoneu que també podeu utilitzar-los per recórrer els caràcters d'una cadena:

```python
fruit = 'apple'

def traverse_characters2(word):
    new_word = ""
    for i in range(0, len(word)):
        new_word += word[i] + " - "
    return new_word       
    
def traverse_characters3(word):                            
    new_word = ""
    i = 0
    while i < len(word):
        new_word += word[i] + " - "
        i += 1
    return new_word

print(traverse_characters2(fruit)+"\n"+traverse_characters3(fruit))
```

### Extended slices

Slices (subcadenes) a Python poden tenir un tercer argument, que és la mida del pas (o "passada") que es pren entre els índexs. És similar al tercer argument de la funció `range()`. Aleshores, el format dels slices passa a ser `<string>[<begin>:<end>:<step>]`. Per defecte, la mida del pas és 1.

L'ús més habitual de la mida del pas és utilitzar una mida de pas negativa per crear una versió invertida d'una cadena.

```python
fruit = "banana"
print(fruit[::2])
print(fruit[1::2])
print(fruit[::-1]) 
print(fruit[::-2]) 
```

Invertir una cadena utilitzant `[::-1]` és conceptualment similar a recórrer la cadena des de l'últim caràcter fins al començament de la cadena utilitzant passos enrere de mida 1.

```python
fruit = "banana"                
for i in range(len(fruit), -1):
    print(fruit[i])   
```

### Strings are immutable

Una propietat bàsica de les cadenes és que són *immutables*. Això vol dir que no es poden canviar. Per exemple, no podeu canviar un caràcter d'una cadena assignant-li un valor nou. Com a demostració, el codi següent provoca un error d'execució si intenteu executar-lo:

```python
fruit = "oringe"
fruit[2] = "a"
print(fruit)
```

Si voleu fer un canvi a una cadena, heu de crear una nova cadena que contingui el canvi; llavors podeu assignar la cadena nova a la variable existent si voleu. Per exemple:

```python
fruit = "oringe"
fruit = fruit[:2] + "a" + fruit[3:]
print(fruit)
```

Les raons per les quals les cadenes són immutables estan fora de l'abast d'aquest curs. Només recordeu que si voleu modificar una cadena, heu de sobreescriure la cadena sencera i no podeu modificar índexs individuals.

### `string` methods

Hi ha una col·lecció de mètodes dissenyats per funcionar amb cadenes. Tots aquests mètodes s'apliquen a una cadena per realitzar alguna operació. Com que les cadenes són immutables, *mai canvien* la cadena en què treballen, però sempre "retornen" una versió modificada de la cadena.

Tots aquests mètodes s'anomenen com a `<string>.<method>()`, és a dir, heu d'escriure la cadena en què treballen abans de la crida del mètode, amb un punt entremig. Ho trobareu més sovint, i per què s'implementa d'aquesta manera s'explicarà més endavant al curs, en els capítols sobre l'orientació d'objectes.

La majoria d'aquests mètodes no formen part d'un mòdul específic, però es poden cridar sense importar-los. Hi ha un mòdul `string` que conté constants i mètodes específics que es poden utilitzar als vostres programes, però els mètodes que comento aquí es poden utilitzar sense importar el mòdul `string`.

### `strip()`

`strip()` elimina d'una cadena els espais inicials i finals, incloses les noves línies inicials i finals i altres caràcters que es poden veure com a espais. No hi ha paràmetres. Vegeu l'exemple següent (la cadena està vorejada per `[` i `]` per mostrar l'efecte):

```python
s = "    And now for something completely different\n     "
print("["+s+"]")
s = s.strip()
print("["+s+"]")
```

### `upper()` and `lower()`

`upper()` crea una versió d'una cadena de la qual totes les lletres són majúscules. `lower()` és equivalent, però només utilitza lletres minúscules. Cap dels dos mètodes utilitza paràmetres.

```python
s = "The Meaning of Life"
print(s)
print(s.upper())
print(s.lower())
```

### `find()`

`find()` es pot utilitzar per cercar en una cadena l'índex inicial d'una subcadena particular. Com a paràmetres obté la subcadena i, opcionalment, un índex inicial per cercar i un índex final. Retorna l'índex més baix on comença la subcadena, o `-1` si no es troba la subcadena.

```python
s = "Humpty Dumpty sat on the wall"
print(s.find("sat"))
print(s.find("t"))
print(s.find("t", 12))
print(s.find("q"))

s.find(" ")
```

### `replace()`

`replace()` substitueix totes les ocurrències d'una subcadena per una altra subcadena. Com a paràmetres, obté la subcadena a buscar i la subcadena per substituir-la. Opcionalment, obté un paràmetre que indica el nombre màxim de substitucions a realitzar.

He de tornar a subratllar que les cadenes són immutables, de manera que la funció `replace()` no està canviant la cadena. Retorna una cadena nova que és una còpia de la cadena amb les substitucions fetes.

```python
s = ' Humpty Dumpty sat on the wall '
new_s = s.replace('sat on', 'fell off') 

print(new_s)
print(s)
```

### `split()`

`split()` divideix una cadena en paraules, basant-se en un caràcter o subcadena donat que s'utilitza com a separador. El separador es dóna com a paràmetre i, si no es dóna cap separador, s'utilitza l'espai en blanc, és a dir, es divideix una cadena en les paraules reals (tot i que la puntuació adjunta a les paraules es considera part de les paraules). Si hi ha diverses ocurrències del separador una al costat de l'altra, s'ignoren les addicionals (és a dir, amb l'espai en blanc com a separador, no importa si hi ha un únic espai en blanc entre dues paraules o múltiples).

El resultat d'aquesta divisió és una anomenada "llista" de cadenes. Les llistes es discuteixen en un proper capítol. Tanmateix, tingueu en compte que si voleu accedir a les paraules separades, podeu utilitzar la construcció `for <word> a <list>:`.

```python
s = 'Humpty Dumpty sat on the wall'
wordlist = s.split()
for i in wordlist:
    print(i)
print(wordlist)
```

Una propietat molt útil de la divisió és que podem descodificar alguns formats de fitxer bàsics. Per exemple, un fitxer de valors separats per comes (CSV) és un format molt senzill, del qual la configuració bàsica és que cada línia consta de valors separats per comes. Aquests valors es poden dividir entre ells mitjançant el mètode `split()`. (Nota: en realitat serà una mica més complicat, ja que pot haver-hi comes als camps que s'emmagatzemen al fitxer CSV, de manera que depèn una mica del contingut del fitxer si aquest enfocament senzill funcionarà. Més informació sobre fitxers CSV es dirà en un capítol posterior del curs, on es comenten els formats dels fitxers.)

```python
csv = "2016,September,28,Data Processing,Tilburg University,Tilburg"
values = csv.split(',')
for value in values:
    print(value)

print("")
print(values)
print (values[1][0])
```

### `join()`

`join()` és el contrari de `split()`. `join()` uneix una llista de paraules, separades per un separador específic. Sembla que seria un mètode de llistes, però per raons històriques es defineix com un mètode de cadena. Com que tots els mètodes de cadena es criden amb el format `<string>.<method>()`, hi ha d'haver una cadena davant la crida a `join()`. Aquesta cadena és el separador que voleu utilitzar, mentre que el paràmetre del mètode és la llista que voleu unir. El valor de retorn, com sempre, és la cadena resultant. A l'exemple següent, tingueu en compte la notació de cadascun d'aquests passos:

```python
s = "Humpty;Dumpty;sat;on;the;wall"
print (s)
wordlist = s.split(';')
print (wordlist)
s = " ".join(wordlist)
print(s)
```

```python
def extract_word(text):
    words = []

    index = text.find("[")

    while index != -1:
        word_start = index + 1
        word_end = text.find("]", index)

        word = text[word_start:word_end]

        words.append(word)

        index = text.find("[", index + 1)

    print(words)
    
extract_word(text)
```

**Exercise 6.4 (optional):** In the text below, count how often the word "wood" occurs (using program code, of course). Capitals and lower case letters may both be used, and you have to consider that the word "wood" should be a separate word, and not part of another word. Hint: If you did the exercises from this chapter, you already developed a function that "cleans" a text. Combining that function with the `split()` function more or less solves the problem for you.

```python
# Counting wood.
text = """How much wood would a woodchuck chuck
If a woodchuck could chuck wood?
He would chuck, he would, as much as he could,
And chuck as much as a woodchuck would
If a woodchuck could chuck wood."""

```

```python
text = text.lower()
text = text.replace("?", "")
text = text.replace(",", "")
text = text.replace(".", "")
text = text.replace("/n", " ")
text = text.split()

def find_wood(text):
    wood_c = 0
    for word in text:
        if word == "wood":
            wood_c += 1   
    return wood_c

find_wood(text)
```
