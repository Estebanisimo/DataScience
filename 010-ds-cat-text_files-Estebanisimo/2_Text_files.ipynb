## Fitxers de text pla

Quan els programadors es refereixen a "fitxers de text" o "fitxers de text pla", es refereixen a fitxers en què tots els caràcters s'han de llegir com a caràcters normals, com escriuria amb un teclat. Per exemple, els fitxers de programa Python són fitxers de text pla, igual que els fitxers HTML. Els documents del processador de textos, però, no són fitxers de text plans, ni tampoc les imatges. Si voleu saber si un fitxer és un fitxer de text o no, podeu provar d'obrir-lo en un editor de text (com l'editor de l'entorn IDLE, que ve amb Python). Si només veieu text llegible, és probable que el fitxer sigui un fitxer de text. En cas contrari, és un anomenat "fitxer binari".


Els fitxers de text consisteixen en línies de text. Al final d'una línia, hi ha un símbol de "nova línia", que en Python és el caràcter "`\n`". Els diferents sistemes operatius utilitzen maneres lleugerament diferents d'emmagatzemar aquest caràcter en un fitxer de text: alguns programes de Windows l'emmagatzemen com a "retorn de carro", així com com a "alimentació de línia" ("`\r\n`", de vegades pensat com un sobrant de antigues màquines d'escriure), mentre que a Linux sempre s'emmagatzema com un sol "`\n`". Sempre que accediu a un fitxer des de Python com a fitxer de text normal, Python convertirà els caràcters que llegeix a l'estàndard "`\n`", i viceversa quan escrigui. Per tant, normalment no us haureu de preocupar per aquestes diferències (excepte quan necessiteu transferir fitxers de text entre sistemes operatius).

Malgrat això; si obriu un fitxer en altres editors, com ara el bloc de notes, de vegades no entén correctament els caràcters de final de línia i ho mostra tot en una línia.

### Controladors i punters de fitxers (File handles and pointers)

Quan treballeu amb un fitxer en un programa, heu d'obrir-lo. L'obertura d'un fitxer proporciona l'anomenat "control de fitxer". Un identificador de fitxer es pot veure com un punt d'accés al fitxer. Conté un "punter" que indica un lloc concret del fitxer. Aquest punter s'utilitza quan llegiu o escriviu al fitxer. Per exemple, quan llegiu el fitxer, comença a llegir-lo al punter i mou el punter cap endavant al fitxer.

Quan obriu un fitxer, **el punter es col·loca en un lloc concret del fitxer**, depenent de com heu obert el fitxer.

* Si heu obert el fitxer només per a la lectura, el punter es col·loca al començament del fitxer. 
* El mateix passa quan obriu el fitxer tant per llegir com per escriure. 
* Si obriu el fitxer per "afegir" (és a dir, per col·locar dades noves al final del fitxer), el punter del fitxer es col·locarà al final del fitxer. 
* Finalment, si obriu un fitxer només per escriure, en realitat el fitxer es buida completament i el punter del fitxer es col·loca a l'inici del fitxer, ara buit. **texto en negrita**

Per crear un fitxer nou (és a dir, un fitxer amb un nom que encara no existeix), l'obriu per "només per escriure".

Després d'obrir el fitxer, el controlador del fitxer és l'únic punt d'accés al fitxer. Totes les accions que realitzeu al fitxer, les feu com a mètodes per al controlador del fitxer.

**Tingueu en compte que qualsevol sistema operatiu només permet obrir un nombre limitat de fitxers simultàniament. Per tant, hauríeu de tancar els fitxers amb els quals ja no necessiteu treballar.**

### Moure el punter del fitxer

El punter del fitxer, que indica on en un fitxer esteu treballant, es mou automàticament. Per exemple, quan llegiu 10 caràcters d'un fitxer, el punter del fitxer indica quin és el primer d'aquests 10 caràcters i, mentre llegeix, es mou 10 caràcters cap amunt, de manera que la seva nova posició és 10 caràcters més enllà del fitxer que abans. Quan tracteu amb fitxers de text, els moviments automàtics del punter del fitxer són exactament el que voleu. És possible col·locar el punter del fitxer manualment mitjançant mètodes específics, però aquests mètodes, en general, només s'utilitzen quan es tracten fitxers binaris.

### Emmagatzematge en memòria intermèdia (Buffering)

Quan feu canvis als fitxers, sovint no s'emmagatzemen als fitxers immediatament. En comptes d'això, el sistema operatiu **"emmagatzema"** els canvis a la memòria i "bufer" els buffers als fitxers reals quan veu la necessitat d'això. Podeu forçar la neteja dels buffers tancant un fitxer. Els buffers també s'esborren quan el programa acaba amb normalitat.

Tanmateix, quan el vostre programa falla (per exemple, a causa d'un error en temps d'execució), és possible que els buffers no s'esborrin i els vostres fitxers no s'actualitzaran fins al punt on es va produir l'error. Per tant, no podeu tenir en compte el contingut del fitxer quan intenteu depurar un programa.

### Programes de processament de fitxers

La majoria de programes que tracten fitxers de text segueixen un procés que, en bucle:

* llegeix el contingut d'un fitxer, 
* processa aquests continguts d'alguna manera 
* i després escriu el contingut en un altre fitxer. 

Per exemple, un programa pot llegir línies d'un fitxer de text i, per a cada línia, ordenar les paraules, després escriure les paraules ordenades en un altre fitxer de text. Això no és gaire diferent d'un programa que demana a l'usuari que proporcioni, en un bucle, una línia de text, després ordena les paraules a la línia i les mostra amb la funció `print()`.

Tanmateix, acostuma a ser experimentat com més complicat.

Tot i que treballar amb fitxers pot donar una sensació de falta de control, durant el desenvolupament del programa sempre podeu incloure declaracions `print()` per obtenir una visió del que està fent el programa. Per exemple, quan llegeix una línia, podeu imprimir aquesta línia, i quan escriu una línia, també podeu imprimir aquesta línia. D'aquesta manera, la vostra visió del funcionament intern del programa no és diferent, independentment de si feu servir entrades manuals i sortides de pantalla, o entrades i sortides de fitxers.

## Llegint fitxers de text

To read the contents of a file, you must first open it, then read the contents, then close it.

### Opening a file using `open()`

Per obrir un fitxer, feu servir la funció `open()`.

La funció `open()` obté dos arguments, dels quals el segon és opcional. El primer argument és el nom del fitxer. Si el fitxer no es troba al directori actual, heu d'incloure la ruta completa del fitxer perquè Python el pugui trobar. El segon argument és el "mode". El mode indica com voleu tractar el fitxer. El mode predeterminat (que s'escull quan no proporcioneu el segon argument) és obrir el fitxer com a fitxer de text només per a la lectura. Com establir altres modes es comentarà més endavant.

La funció `open()` retorna un identificador de fitxer, que feu servir per a totes les funcionalitats restants.

En lloc d'escriure "`<handle> = open( <filename> )`", sovint veureu programes Python que l'escriuen com "`open( <filename> ) as <handle>`". El segon mètode té un avantatge del que parlarem més endavant

### Llegint un fitxer amb `read()`

La manera més senzilla de llegir el contingut d'un fitxer és utilitzar el mètode `read()`, sense arguments, al controlador del fitxer (handle). Això retorna una cadena que conté el contingut complet del fitxer. `read()` pot obtenir un argument, però només s'utilitza per a fitxers binaris.

La lectura d'un fitxer mou el punter del fitxer just després de la part del fitxer que s'ha llegit. Això vol dir que si utilitzeu el mètode `read()` sense arguments, el punter del fitxer es mou al final del fitxer. Això implica que si intenteu fer `read()` per segona vegada, no es llegiria res, ja que no hi ha res a llegir després del punt on es troba el punter del fitxer.

```python
filepath = f"{CURR_DIR}/my_text.txt"
f = open(filepath, "r") # handler
print(f.read())
```

### Tancar un fitxer amb `close()`

Per tancar un fitxer, utilitzeu el mètode `close()` al controlador del fitxer. Cada fitxer que obriu, hauríeu de tancar-vos en algun moment del vostre programa. Això és especialment cert en aquests quaderns, ja que cada pàgina de quadern es considera un programa complet; això vol dir que si no tanqueu un fitxer, romandrà obert fins i tot després que finalitzi el fragment de codi que heu executat.

Si tot el que heu de fer amb un fitxer es fa en un sol bloc, podeu escriure aquest bloc de la següent manera:

     with open( <nom del fitxer> ) as <handle>:
         <declaracions>
        
Aquesta construcció sintàctica té l'avantatge que el fitxer es tancarà automàticament després que el bloc `<statements>` finalitzi, de manera que no cal que incloguis una crida explícita a `close()`. Aquesta construcció és típicament Python; no ho veus en molts altres llenguatges de programació.

```python
with open(filepath, "r") as f:
  print(f.read())
```

### Llegint línies utilitzant `readline()`

---



Per llegir un fitxer de text línia per línia, podeu utilitzar el mètode `readline()`. El mètode `readline()` llegeix caràcters que comencen al punter del fitxer fins al següent caràcter de "nova línia" (inclòs), i els retorna com a cadena. Podeu reconèixer que heu arribat al final del fitxer pel fet que ja no es llegeix cap caràcter, és a dir, la cadena que es retorna està buida i per tant hem arribat al final de l'arxiu.

```python
def display_contents_file2(file_name):
    fp = open( file_name )
    while True:
        buffer = fp.readline()             
        if buffer == "":
            break     
        print( buffer )
    fp.close()

display_contents_file2(filepath)
```

Tingueu en compte que la sortida del codi anterior té una línia buida entre cadascuna de les línies que es mostren. D'on ve aquesta línia addicional? Pensa-hi.

La línia addicional hi és perquè el mètode `readline()` retorna una cadena dels caràcters llegits, fins i *incloent* el caràcter de nova línia. Així, quan s'imprimeix el "buffer", també imprimeix un caràcter de nova línia. I com que la funció `print()` *també* es mou a una nova línia després d'executar-la, hi ha una línia buida impresa després de cada línia de text.

Una variació del mètode `readline()` és el mètode `readlines()`. `readlines()` llegeix totes les línies del fitxer i les retorna com una llista de cadenes. Les cadenes inclouen els caràcters de nova línia.

```python
# readlines()
def display_contents_file(file_name):
    #with open( file_name ) as fp:
    fp = open(file_name, "r")
    buffer = fp.readlines()
    for line in buffer:
        print(line, end="")
    print(buffer)
    fp.close()

display_contents_file(filepath)
```

### Quan i quin mètode de lectura de fitxers utilitzar

Tant el mètode `read()` com `readlines()` llegeixen tot un fitxer alhora. Òbviament, això és acceptable per a fitxers petits, però per a fitxers llargs és possible que no tingueu prou memòria per emmagatzemar el contingut del fitxer de manera eficient. En aquestes circumstàncies (o quan no coneixeu la mida del fitxer), hauríeu de llegir un fitxer línia per línia amb el mètode `readline()`.

Sovint és una bona idea, durant el desenvolupament del codi, processar només les primeres línies d'un fitxer. D'aquesta manera limiteu la quantitat de temps que necessita el programa per processar un fitxer i limitar la seva sortida, cosa que facilita la depuració. Per exemple, el codi següent processa les 10 primeres línies d'un dels fitxers més llargs.

```python
def display_contents_file4(file_name):
    fp = open(file_name)
    count = 0
    while count < 10:
        buffer = fp.readline()
        if buffer == "":
            break
        print( buffer, end="" )
        count += 1
    fp.close()

display_contents_file4( filepath)
```

Un cop finalitzat el programa i depurat, es poden eliminar les referències a `count` i canviar el bucle a `while True`, per processar tot el fitxer.

## Sistema de fitxers 

El sistema de fitxers d'un ordinador consta d'una organització estructurada en forma d'arbre de directoris i fitxers.



Hi ha un directori "arrel", que és el punt d'accés principal per a tots els altres directoris.
El directori arrel s'identifica amb una barra inclinada (`/`) o una barra invertida (`\`), depenent del sistema operatiu. A Windows és una barra invertida, a Mac OS i Linux és una barra inclinada.

Tanmateix, ara Windows també admet la barra inclinada. És recomanable fer servir la barra inclinada en la majoria dels casos, ja que en les cadenes la barra invertida indica un símbol especial, de manera que si voleu utilitzar una barra invertida en una cadena com a separador de directoris, heu d'utilitzar una barra inclinada doble invertida (per poder escapar el seu ús reservat).
Això acostuma a ser confús, per això us recomano utilitzar la barra inclinada.

"Sota" el directori arrel hi ha molts altres directoris, cadascun identificat amb un nom, i normalment també diversos fitxers, cadascun identificat amb un nom. Sota cada directori hi pot haver més directoris i fitxers.


Cada sistema operatiu té certes restriccions sobre quins noms de fitxers i directoris es poden utilitzar, però en general s'admeten la majoria de caràcters. És convenció que els fitxers normals tenen una extensió, que es col·loca al final del nom del fitxer i separada del nom del fitxer.
amb un punt. 

L'extensió identifica quin tipus de fitxer és, per exemple:

* un programa executable (.exe), 
* un fitxer de text pla (.txt) o 
* un fitxer Python (.py). 

També és convenció que els noms de directoris no tinguin aquesta extensió. Tanmateix, això no és una regla, i és possible que trobeu fitxers sense extensió i directoris amb extensió. **Tingueu en compte que a l'entorn visual, les extensions dels fitxers sovint s'oculten, però hi són; simplement no les veieu.**

**Per identificar de manera única un fitxer, cal conèixer el seu "camí o ruta" exacte des de l'arrel fins al fitxer, seguint els directoris.**

El nom de la ruta del fitxer és: 

> `/<directori>/<directori>/.../<nom del fitxer>`.

A Windows, es pot col·locar una lletra d'unitat davant d'aquest ruta, fent-la 

> `<unitat>:/<directori>/<directori>/.../<nom del fitxer>`.

Per exemple, si a Windows, a la unitat "`C`", sota l'arrel hi ha un directori "`Python34`", sota el qual hi ha un directori "`Lib`", en el qual podeu trobar un fitxer "`os.py`", la ruta per a aquest fitxer és `C:/Python34/Lib/os.py`.

**A Windows, aquesta ruta no distingeix entre majúscules i minúscules, de manera que només podeu utilitzar lletres minúscules si voleu. Això no és el cas de tots els sistemes operatius, però.**

Quan esteu treballant al sistema de fitxers (i sempre esteu treballant al sistema de fitxers, encara que no us n'adoneu), hi ha un "directori actual", que s'identifica amb un punt (.). Si voleu accedir a un fitxer del directori actual, no cal que conegueu la ruta complet; n'hi ha prou amb saber el nom del fitxer. Un directori "superior" al directori actual (és a dir, el directori principal) s'identifica amb un punt doble (..). El directori principal de l'arrel és l'arrel mateixa.

## Mètodes d'`os.path`

En aquest punt ja sabeu tot el que necessiteu per gestionar fitxers de text a Python. Tanmateix, hi ha diverses funcions útils que us faciliten la vida quan tracteu fitxers. Aquests es recullen al mòdul `os.path`. Com és habitual, no els enumerarem tots, però sí els que més utilitzareu.

En aquestes funcions, el terme "ruta(path)" es refereix a un nom de fitxer o de directori, amb els directoris principals (i la lletra de la unitat). Els directoris pares (i la lletra de la unitat) no *necessiten* estar-hi explícitament, però encara que no ho siguin, implícitament encara ho són, ja que cada fitxer i cada directori es troba en un lloc determinat del sistema de fitxers.

### `exists()`

La funció `exists()` obté una ruta com a argument i retorna `True` si aquest ruta existeix, i `False` si no.

```python
from os.path import exists

def check_if_path_exists(file_name):
    if exists(file_name ):
        return( file_name + " exists" )
    else:
        return( file_name + " does not exist" )

print(check_if_path_exists("Shakespeare.txt")) # not in current directory
print(check_if_path_exists("my_text.txt")) # not in current directory

exists(f"{CURR_DIR}/Shakespeare.txt") # it is in our handcrafted path
```

### `isfile()`

`isfile()` prova si la ruta que s'ofereix com a argument és un fitxer. Si ho és, retorna "True". Si no ho és, retorna "False". Si la ruta no existeix, la funció també retorna `False`

```python
from os.path import isfile

def check_if_file_exists(file_name):
    if isfile( file_name ):
        return( file_name + " is a file" )
    else:
        return( file_name + " is not a file" )
    
print(check_if_file_exists("Shakespeare.txt"))
```

```python
print(check_if_file_exists(f"{CURR_DIR}/Shakespeare.txt"))
```

### `isdir()`

`isdir()` prova si la ruta que s'ofereix com a argument és un directori. Si ho és, retorna "True". Si no ho és, retorna "False". Si la ruta no existeix, la funció també retorna `False`.

```python
from os.path import isdir

def check_if_path_is_dir(file_name):
    if isdir( file_name):
        print( file_name + " Sheakspeare is a directory" )
    else:
        print(file_name + " is not a directory" )
        
check_if_path_is_dir(filepath)
```

### `join()`

`join()` pren una o més parts d'una ruta com a argument i les concatena de manera raonablement intel·ligent a un nom "legal" per a una ruta. Això vol dir que afegirà i eliminarà barres inclinades segons sigui necessari. `join()` és especialment útil en combinació amb `listdir()`.

La raó per la qual `join()` és útil amb `listdir()`, és que `listdir()` dóna com a resultat una llista de noms de fitxers que no inclouen els noms dels directoris. 

Normalment, quan demaneu una llista de noms de fitxers, teniu la intenció d'obrir-los en algun moment. Però per obrir un fitxer que no es troba al directori actual, heu de conèixer el nom complet del camí que porta al fitxer. Quan apliqueu `listdir()`, sabeu on esteu buscant els fitxers, de manera que coneixeu els elements del nom del camí. 

Per construir el nom de camí complet per a cada fitxer, heu de concatenar els elements del nom del camí amb el nom del fitxer. En lloc d'intentar decidir on heu d'afegir barres inclinades i quin tipus de barres han de ser, podeu deixar-ho tot a la funció `join()`.

El codi següent cerca tots els fitxers del directori actual i els enumera inclòs el seu nom de camí complet. Vegeu com s'utilitza `join()` per construir aquest nom de camí des del directori actual i el nom del fitxer

```python
from os import listdir , getcwd
from os.path import join

def print_paths(dir_path):
  filelist = listdir( dir_path )

  if dir_path == ".":
    dir_path = getcwd()
 
  for name in filelist:
    pathname = join( dir_path, name )
  print( pathname )

print_paths(".")
```

```python
print_paths(dirname(filepath))
```

## Codificació de fitxers

Els fitxers de text utilitzen una "codificació", és a dir, un sistema que prescriu com se suposa que s'han d'interpretar els caràcters dels fitxers. Aquesta codificació pot ser diferent entre els sistemes operatius. Podeu veure la codificació preferida que fa servir el vostre sistema amb una crida a `sys.getfilesystemencoding()`.

```python
from sys import getfilesystemencoding

print( getfilesystemencoding() )
```

Si llegiu un fitxer de text que utilitza una codificació diferent de la que prefereix el vostre sistema de fitxers, és possible que obtingueu un "UnicodeDecodeError". Si rebeu o no aquest error per a un fitxer en concret, està relacionat amb el vostre sistema operatiu. Una conseqüència molesta d'això és que quan porteu el codi Python que llegeix un fitxer a un altre sistema, un fitxer que el vostre codi podria llegir prèviament pot provocar que el vostre codi es bloquegi després del port.

Una manera senzilla d'evitar aquest problema és afegint un paràmetre addicional en obrir un fitxer, que indica el mecanisme de codificació que voleu utilitzar en llegir el fitxer. Podeu fer-ho afegint un paràmetre "`encoding=<encodingname>`", on `<encodingname>` és una cadena que pot tenir una varietat de valors, per als quals alguns són típics:

    codificació ascii de 7 bits, admet caràcters amb valors en el rang 00-7F
    codificació latin-1 de 8 bits, admet caràcters amb valors en el rang 00-FF
    mbcs de codificació de 2 bytes, que actualment s'està substituint per UTF-8
    codificació de bytes variables utf-8

Normalment, els fitxers de text es creen amb la codificació "ascii" o "latin-1". Com que `ascii` està incorporat a `latin-1`, podeu obrir qualsevol fitxer de text amb seguretat especificant `latin-1` com a codificació. És possible que per als caràcters més enllà de l'interval `ascii`, obtingueu caràcters diferents dels que la persona que va crear el fitxer volia que vegeu, això depèn del mecanisme de codificació que utilitzi el vostre sistema de fitxers. Però almenys s'evita l'`UnicodeDecodeError`.

Tingueu en compte que, tot i que `utf-8` admet un rang de caràcters molt més ampli que `latin-1`, encara podeu obtenir l'`UnicodeDecodeError` quan llegiu un fitxer de text que utilitza la codificació `latin-1` en un sistema que utilitza ` codificació utf-8`, ja que `utf-8` no té caràcters corresponents amb valors en el rang 80-FF.

Si voleu veure quins caràcters especials s'admeten amb valors del rang 80-FF al vostre sistema, executeu el codi següent. El valor numèric d'un caràcter de la taula es pot derivar calculant `16*fila+col`, per la qual cosa `fila` i `col` són el número hexadecimal de fila i columna, respectivament. No mostro els caràcters del rang 80-9F, ja que normalment no s'emplenen.

```python
for i in range(16):
    if i < 10:
        print( ' '+chr( ord( '0' )+i ), end='' )
    else:
        print( ' '+chr( ord( 'A' )+i-10 ), end='' )
print()
for i in range( 10, 16 ):
    print( chr( ord( 'A' )+i-10 ), end='' )
    for j in range( 16 ):
        c = i*16+j
        print( ' '+chr( c ), end='' )
    print()
```

### Exercici 5

Escriu un programa que llegeixi el contingut del fitxer "blakepoems.txt", el divideixi en paraules (on tot el que no és una lletra es considera un límit de paraula) i crea un diccionari sense distinció de majúscules i minúscules que emmagatzema per a cada paraula amb quina freqüència es fa. es produeix al text. A continuació, imprimiu totes les paraules amb les seves quantitats en ordre alfabètic.

```python
# Counting words in blakepoems.txt.
fp = open( "blakepoems.txt")

import re

from os.path import join

def clean( s ):
  news = ""
  s = s.lower()
  for c in s:
    if c >= "a" and c <= "z":
      news += c
    else:
      news += " "
  return news

filepath = join(CURR_DIR,"blakepoems.txt")
fp = open( filepath )

text = fp.read()
fp.close()

wdict = {}
for word in clean( text ).split():
  wdict[word] = wdict.get( word , 0 ) + 1

keylist = list( wdict.keys() )
keylist.sort()
for key in keylist:
  print( "{}: {}".format( key, wdict[key] ) )
```

Feu el mateix que vau fer a l'exercici anterior, però ara processeu el text línia per línia. Això és una cosa que hauríeu de fer si haguéssiu de processar un text molt llarg (com és el cas).

```python
# Counting words line by line.

wdict = {}
fp = open( filepath )
while True:
  line = fp.readline()
  if line == "":
    break
  for word in clean( line ).split():
    wdict[word] = wdict.get( word , 0 ) + 1
fp.close()
keylist = list( wdict.keys() )
keylist.sort()
for key in keylist:
  print( "{}: {}".format( key, wdict[key] ) )
```

### Exercici 7 (Opcional)

En aquest directori trobareu un fitxer `blakepoems.txt`. Escriu un programa que processi el contingut d'aquest fitxer, línia per línia. Crea un fitxer de sortida al directori de treball actual anomenat `blkpms.txt`, que té el mateix contingut que `blakepoems.txt`, excepte que s'eliminen totes les vocals (insensible a majúscules i minúscules). Al final, mostra quants caràcters has llegit i quants caràcters has escrit. Si voleu comprovar el contingut de `blkpms.txt`, podeu obrir-lo en un editor de text o mostrar les primeres 10 línies més o menys al final del vostre programa.

```python
# Removing vowels.

from os.path import join

def removevowels( line ):
  newline = ""
  for c in line:
    if c not in "aeiouAEIOU":
      newline += c
  return newline

inputname = join( CURR_DIR, "blakepoems.txt" )
outputname = join( CURR_DIR, "blkpms.txt" )

fpi = open( inputname )
fpo = open( outputname , "w" )

countread = 0
countwritten = 0

while True:
  line = fpi.readline()
  if line == "":
    break
  countread += len( line )
  line = removevowels( line )
  fpo.write( line )
  countwritten += len( line )

fpo.close()
fpi.close()

print( "Characters read:", countread )
print( "Characters written:", countwritten )
```
