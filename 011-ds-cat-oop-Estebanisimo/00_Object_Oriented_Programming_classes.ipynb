## El món orientat a objectes

imagina't assegut a la taula de la teva cuina. Al teu costat hi ha un bol de fruita amb unes pomes. Tot i que aquestes pomes comparteixen certes característiques, també tenen les seves diferències. Comparteixen el seu nom, el seu preu i la seva "edat", però tots tenen pesos (lleugerament) diferents. També hi ha algunes taronges al bol que, com les pomes, són fruites, però difereixen en altres característiques : diferents noms, diferents colors, diferents arbres dels que creixen...

Es a dir, comparteixen **atributs** amb les pomes  que també comparteixen amb altres tipus de fruites però que alhora les fa diferents de, per exemple, la cadira on estàs assegut. Puc menjar-me una fruita, pomes, taronges... No intentaré menjar-me una taula.

Quan intentem modelar el nostre món en un programa informàtic, hem de modelar objectes: objectes com pomes, taronges i taules. Alguns d'aquests objectes tenen *molt* en comú, per exemple, **cada poma comparteix moltes característiques amb totes les altres pomes**. 

Aleshores, és dessitjable definir una **classe** "poma" que contingui les característiques que comparteixen totes les pomes i només omplir les poques característiques en què les pomes difereixen entre si per a cada objecte de poma individual.

El mateix passa amb les taronges, haurien d'obtenir la seva pròpia classe "taronja". I encara que "pomes" i "taronges" són força diferents, encara comparteixen algunes característiques que impliquen que m'agradaria posar-les en la mateixa classe: la classe "fruita". Cada objecte que pertany a la classe "fruita" almenys té la propietat de poder menjar-lo. El que significa que cada objecte de poma individual no només pertany a la classe "poma", sinó que també pertany a la classe "fruita", igual que les "taronges".

Això condueix a que si intento modelar el món, o una part del món, necessito modelar objectes, i en lloc de modelar cada objecte per separat, **és millor definir classes d'objectes**, ja que això vol dir que puc fer declaracions sobre determinats grups d'objectes en general. ### Estudiants, professors i cursos

Molts programes tracten amb persones. L'administració d'estudiants tracta amb els estudiants, que són persones. Aquests alumnes segueixen cursos, que són impartits per professors, que també són persones. Sens dubte, l'administració d'estudiants emmagatzema informació sobre estudiants i professors, i probablement el programador que va crear el programari per a l'administració d'estudiants va ser prou intel·ligent com per crear una única interfície que permeti introduir dades de la persona.

Quines dades comparteixen totes les persones, pel que fa a l'administració dels estudiants? Bé, probablement totes les persones tenen un nom i un cognom. Tenen una adreça. També tenen una edat i un gènere. A tots se'ls assigna un número d'administració, de manera que per a l'administració tenen almenys una cosa que els fa únics. Tots aquests elements de dades són "propietats" o "atributs" de "persones".


> Hem esmentat les propietats nom, cognoms, adreça, edat, sexe i número d'administració. Un d'ells s'assembla més a una funció que a una propietat. Veus quin?


La resposta és "edat". L'edat es calcula a partir de la data de naixement i la data actual. Tot i que podeu considerar l'edat una propietat, és una propietat que s'ha de calcular cada vegada que es necessiti. No podeu emmagatzemar-lo com a valor, ja que demà pot ser diferent d'avui, sense que canviï res més que la data. Per tant, si dissenyo una classe "Persona" que modeli una persona, millor que la "data de naixement" sigui un atribut de la persona, mentre que "l'edat" és un mètode de la persona. 

> Recordeu que els mètodes són funcions que pertanyen a un tipus de dades determinat: si es defineix un tipus de dades `Persona`, `data_de_naixement` és un atribut d'aquest, mentre que `edat()` sería un mètode que retorna l'edat de la persona com a nombre enter utilitzant l'atribut como a entrada la `data_de_naixement`.

Els estudiants i els professors són tots dos persones. Comparteixen les propietats de la classe `Persona`. No obstant això, hi ha diferències. Els professors, per exemple, reben un sou, mentre que els estudiants no. Els estudiants, en canvi, guanyen notes als cursos, cosa que els professors no; imparteixen els cursos. D'això se'n deriven dues observacions òbvies:

- Si bé els estudiants i els professors són persones, tenen clares diferències; a més d'una classe "Persona", necessito una classe "Estudiant" i una classe "Professor", ambdues derivades de la classe "Persona".
- Els "cursos" semblen ser una part inherent del món de l'administració dels estudiants, per la qual cosa també pot ser necessària una classe "Curs".

Un cop el `Curs` s'ha convertit en una classe en el món de l'administració dels estudiants, les relacions es fan visibles. Els estudiants tenen relacions amb diversos cursos, i els professors també, encara que amb una capacitat diferent. Els estudiants "s'inscriuen" als cursos. Sembla que es necessita un mètode `enroll()`, que permeti a un estudiant relacionar-se amb un curs. La pregunta és: `enroll()` és un mètode de `Estudiant`, que rep un curs com a argument, o és un mètode de `Curs`, que obté un estudiant com a argument? Què penses?

> La resposta és: "depèn". Depèn de com us imagineu el món de l'administració dels estudiants. Per a mi, em sembla més natural fer de `enroll()` un mètode d'un curs, ja que veig un curs com una col·lecció d'estudiants. Tanmateix, en principi no hi ha res en contra de veure un estudiant com una entitat que engloba una col·lecció de cursos. També podeu decidir fer `enroll()` un mètode de cadascun d'ells, o pensar en una altra classe que contingui el mètode `enroll()` que tingui com a arguments l'estudiant i el curs.

Això il·lustra la dificultat de la visió orientada a objectes sobre el disseny del programa: en dissenyar les classes que formen el model mundial amb el qual treballa el programa, s'han de prendre decisions que poden tenir un gran impacte en el funcionament del programa. Les males decisions poden comportar dificultats en el futúr durant la implementació. Heu de dedicar un temps considerable a dissenyar el model orientat a objectes que subjau al programa i intentar preveure totes les conseqüències de les vostres eleccions. Això és difícil fins i tot per als dissenyadors experimentats. Tanmateix, un model sòlid orientat a objectes fa que els programes siguin fàcils de llegir, comprendre i mantenir. El paradigma orientat a objectes sovint val la pena la molèstia.

### Classes, objectes i jerarquies

En el món orientat a objectes, cada entitat distingible pertany a una "classe". Una classe és un model general per a un grup específic d'entitats. Descriu tots els atributs que tenen aquestes entitats i descriu els mètodes que la classe ofereix al món exterior per influir-hi.

Una classe, per si mateixa, *no* és una entitat. **Una entitat que pertany a la classe, és un "objecte". La terminologia és que un objecte és una "instància" d'una classe en particular. Mentre que la classe descriu els seus atributs, un objecte que és una instància de la classe té valors per a aquests atributs.** Tot i que la classe descriu els mètodes que admet, per executar aquest mètode cal un objecte que sigui una instància de la classe amb la qual cridar el mètode.

Una classe és un tipus de dades, un objecte és un valor.

Les classes poden existir en jerarquies. Una classe general d'alt nivell pot descriure propietats i mètodes que comparteixen diferents subclasses. Cada subclasse pot afegir propietats, afegir mètodes i fins i tot canviar propietats i mètodes (tot i que en general no poden -- i no hauria -- eliminar-los completament). Cada subclasse pot tenir més subclasses.

Per exemple, la classe "Apple" pot ser una subclasse de la classe "Fruits", que pot ser una subclasse de la classe "Aliments". Això vol dir que quan en un programa es necessita un objecte de la classe `Food`, podeu proporcionar un objecte que sigui una instància de la classe `Food`, però també un objecte que sigui una instància de la classe `Fruit`, o un objecte que és una instància de la classe `Apple`. Això no funciona al revés, però. Quan, per exemple, una funció d'un programa s'ha dissenyat per a instàncies de `Apple`, no podeu utilitzar-la amb instàncies de `Fruit` o altres subclasses de `Fruit`. Mentre que una `Apple` és `Fruit`, `Fruit` no és una `Apple` i les `Apple`s no són `Orange`.

Aquesta jerarquia s'implementa mitjançant "herència", que és el que explicarem a la següent classe.

### Classes i tipus de dades en Python

La majoria dels llenguatges de programació orientats a objectes tenen alguns tipus de dades "base" i us permeten crear classes, és a dir, nous tipus de dades. Aquest va ser el cas de Python fins a Python 2. Des de Python 3, cada tipus de dades és una classe.

Podeu reconèixer part d'això per la manera en què moltes funcionalitats dels tipus de dades bàsics s'implementen com a mètodes. Recordeu que un mètode sempre s'anomena `<variable>.<method>()`, al contrari de les funcions que funcionen amb una variable, que s'anomenen `<function>( <variable> )`. El fet que quan vulgueu crear una versió en minúscules d'una cadena, ho feu com a `<string>.lower()` ja indica que la cadena és una instància d'una classe.

Però no només les cadenes són instàncies de classe: els enters i els flotants també ho són. Fins i tot tenen mètodes, encara que poques vegades s'utilitzen de manera explícita. Alguns mètodes s'utilitzen implícitament, per exemple, quan afegiu dos nombres juntament amb `+`, en realitat és una trucada de mètode.

## Classes i objectes

Ara que les filosofies bàsiques de l'orientació a objectes estan fora del camí, parlarem de com utilitzar l'orientació d'objectes a Python. Començem amb la creació de noves classes utilitzant la paraula clau `class`.

### `class`

Una classe es pot considerar un tipus de dades nou. Un cop creada una classe, podeu assignar instàncies de la classe a variables. Per començar de manera senzilla, crearem una classe que representi un punt de l'espai 2D. Anomeno aquesta classe "Punt" (el nom de les classes està restringit als mateixos requisits que el de les variables, i **és convenció que els noms de les classes comencin amb una majúscula**). Crear aquesta classe en Python és increïblement fàcil:

```python
class Point:
    pass
```

La paraula clau `pass` a la definició de la classe significa "no fer res". Aquesta paraula clau es pot utilitzar allà on necessiteu col·locar una declaració, però encara no teniu res per col·locar-hi. Tan bon punt s'afegeixen declaracions, ja no necessiteu "passar".

Per crear un objecte que sigui una instància de la classe, assignem a una variable el nom de la classe, amb parèntesis després, com si es tractés d'una trucada de funció (podem tenir arguments entre parèntesis, més endavant en parlarem).

```python
class Point:
    pass

p = Point()
print( type( p ) )
```

Per descomptat, un punt és més que un objecte. Un punt té una coordenada "x" i una "y". Com que Python és un llenguatge de tipatge suau (soft-typed), hem d'assignar valors als atributs per crear-los. Això es fa amb un mètode d'inicialització especial a la classe.

### `__init__()`

El mètode d'inicialització d'una classe té el nom `__init__` (és a dir, dos guions baixos, seguits de la paraula `init`, seguit de dos guions baixos més). Encara que el mètode `__init__()` no estigui definit explícitament per a la classe, el mètode existeix. Utilitzeu el mètode `__init__()` per inicialitzar tot el que voleu inicialitzar en crear una instància de la classe.

En el cas de `Point`, `__init__()` hauria d'assegurar que qualsevol objecte `Point` tingui una coordenada `x` i una `y`. Això s'implementa de la següent manera:

```python
class Point:
    def __init__( self ):
        self.x = 0.0
        self.y = 0.0

p = Point()
print( "({}, {})".format( p.x, p.y ) )
```

Estudieu de prop el codi anterior.`__init__()` es defineix de la mateixa manera que definiríeu una funció dins de la definició de classe.

`__init__()` obté un paràmetre, que s'anomena `self`. Cada mètode que definiu, *sempre* obté almenys un paràmetre, que s'omplirà amb una referència a l'objecte per al qual es crida el mètode. Per convenció, aquest primer paràmetre sempre s'anomena `self`. Això no és obligatori, però tothom ho fa sempre així. Si oblideu incloure aquest primer paràmetre, obtindreu un error d'execució. Si oblideu incloure el primer paràmetre `self` probablement també obtindreu un error d'execució.

En el mètode `__init__()` per a `Punt`, l'objecte que es crea obté dos atributs, que són variables i que formen part de l'objecte. S'anomenen `x` i `y`, i com que formen part de l'objecte, us referiu a ells com a `self.x` i `self.y`. Tots dos obtenen el valor inicial 0.0, cosa que els fa floats.

Per fer referència a aquests atributs quan s'ha creat l'objecte, utilitzeu la sintaxi `<object>.<attribute>`, com podeu veure a l'última línia del codi, on l'objecte que s'acaba de crear s'utilitza en un `print() ` declaració.

Us podríeu preguntar si només podeu crear atributs per a un objecte amb el mètode `__init__()`. La resposta és: no, també podeu crear atributs en altres mètodes, i fins i tot fora de la definició de classe.

```python
class Point:
    def __init__( self ):
        self.x = 0.0
        self.y = 0.0

p = Point()
p.z = 0.0
print( "({}, {}, {})".format( p.x, p.y, p.z ) )
```

La majoria dels programadors de Python tindrien en compte el que passa al codi anterior. És una bona pràctica crear tots els atributs que necessiteu exclusivament al mètode `__init__()` (tot i que podeu canviar-ne els valors en qualsevol altre lloc), de manera que sàpigues que cada instància de la classe els té, i no hi ha casos més.

Si necessiteu una versió de la classe amb atributs addicionals, podeu utilitzar "herència" per crear noves classes basades en les existents, que sí que tenen aquests atributs addicionals.

Com qualsevol mètode, `__init__()` pot obtenir arguments. Podeu utilitzar aquests arguments per inicialitzar (alguns dels) atributs. Per exemple, si vull crear una instància de "Punt" mentre especifiqueu immediatament els valors de les coordenades "x" i "y", puc utilitzar la següent definició de classe:

```python
class Point:
    def __init__( self, x, y ):
        self.x = x
        self.y = y

p = Point( 3.5, 5.0 )
print( "({}, {})".format( p.x, p.y ) )
```

`__init__()` ara es defineix amb tres paràmetres. El primer segueix sent el `self`, ja que sempre hi ha d'estar. El segon i el tercer s'anomenen `x` i `y`. Aleshores, assignem `x` a `self.x`, i `y` a `self.y`.

Ara anomeno la creació d'un punt amb valors per a les coordenades  `x` i `y` com a arguments. El primer argument es passarà al mètode en el segon paràmetre, i el segon en el tercer paràmetre, ja que el primer paràmetre s'utilitzarà per passar la referència al propi objecte.

Si voleu que sigui opcional que el programador passi aquests valors, podeu donar els valors predeterminats als paràmetres mitjançant una assignació a l'especificació del paràmetre, de la següent manera:

```python
class Point:
    def __init__( self, x=0.0, y=0.0 ):
        self.x = x
        self.y = y

p1 = Point()
print( "({}, {})".format( p1.x, p1.y ) )

p2 = Point( 3.5, 5.0 )
print( "({}, {})".format( p2.x, p2.y ) )
```

### `__repr__()` and `__str__()`

Al codi anterior, imprimim els atributs del punt. Què passa si intento imprimir el punt en si?

```python
class Point:
    def __init__( self, x=0.0, y=0.0 ):
        self.x = x
        self.y = y

p = Point( 3.5, 5.0 )
print( p )
```

Això no és el que volem veure. Quan imprim un punt, volem veure les coordenades. Python ofereix un altre mètode predefinit per a això, que és `__repr__()`. `__repr__()` hauria de retornar una cadena, que conté el que voleu veure quan es mostra un objecte.

```python
class Point:
    def __init__( self, x=0.0, y=0.0 ):
        self.x = x
        self.y = y
    def __repr__( self ):
        return "({}, {})".format( self.x, self.y )
        
p = Point( 3.5, 5.0 )
print( p )
```

Això sembla molt millor.

Python ofereix un altre mètode estàndard per crear una versió de cadena d'un objecte, és a dir, `__str__()`. `__str__()` és el mateix que `__repr__()`, però només s'utilitza quan l'objecte s'està imprimint o es passa a un mètode `format()`. Si `__str__()` no està definit, és el mateix que `__repr__()` (però no viceversa). Si `__str__()` *està* definit, podeu assegurar-vos que quan s'utilitza `print()` es mostra alguna cosa diferent del que es mostra en altres llocs.

Ara podríeu pensar: "quins altres llocs?" L'"altre lloc" principal on es mostren els objectes és a l'intèrpret d'ordres, quan només escriviu el nom de la variable que conté un objecte.

Normalment s'entén que en el mètode `__repr__()` se suposa que heu de retornar una cadena que conté tots i cadascun dels bits d'informació que es necessiten per recrear un objecte, mentre que a `__str__()` podeu tornar una cadena que contingui una representació ben formatada de la informació més important que voleu veure al programa. Molt sovint, aquests dos són iguals.

Molts programadors ignoren `__repr__()` per complet i només defineixen `__str__()`. Sempre hauríeu de definir `__repr__()`, mentre que `__str__()` és opcional. Si utilitzeu `__repr__()`, assegureu-vos que realment retorneu tots els detalls d'un objecte. Si deixeu coses fora, és millor utilitzar només `__str__()`.

## Mètodes

`__init__()`, `__repr__()` i `__str__()` són mètodes predefinits que té cada classe. Tal com els van definir els desenvolupadors de Python, tenen els noms excèntrics que comencen i acaben amb un guió baix doble. Hi ha molts més d'aquests mètodes i es coneixem com "dunder methods" o "magic methods".

També podeu definir els vostres propis mètodes per a una classe. Aquests mètodes obtenen noms semblants als noms que doneu a les funcions i tendeixen a seguir les mateixes convencions: començant amb una lletra minúscula i, si hi ha paraules diferents, o bé tenen guió baix entre elles o bé s'escriu en majúscula la primera lletra de la segona i les paraules posteriors. El prefix `is` s'utilitza per als mètodes que proporcionen una declaració `True`/`False` sobre l'objecte, el prefix `get` s'utilitza per obtenir un valor d'un objecte i el prefix `set` s'utilitza per establir un valor per a un objecte.

Per exemple, per a un punt podem crear un mètode `distance_from_origin()`, que calcula la distància des del punt (0,0) fins al punt donat.

```python
from math import sqrt

class Point:
    def __init__( self, x=0.0, y=0.0 ):
        self.x = x
        self.y = y
    def __repr__( self ):
        return "({}, {})".format( self.x, self.y )
    def distance_from_origin( self ):
        return sqrt( self.x*self.x + self.y*self.y )
        
p = Point( 3.5, 5.0 )
print( p.distance_from_origin() )
```

També podeu crear mètodes que canviïn l'objecte d'alguna manera. Per exemple, la "translació" de punts a una distància es defineix com un desplaçament específic en la direcció horitzontal i vertical. Un mètode `translate()` obté dos arguments (més enllà de la referència `self`), que són els desplaçaments horitzontals i verticals.

### Còpies i referències

A continuació es mostra una còpia del codi anterior, amb unes quantes línies addicionals que fan un canvi al `Punt` `p`.

```python
class Point:
    def __init__( self, x=0.0, y=0.0 ):
        self.x = x
        self.y = y
    def __repr__( self ):
        return "({}, {})".format( self.x, self.y )
        
class Rectangle:
    def __init__( self, point, width, height ):
        self.point = point
        self.width = width
        self.height = height
    def __repr__( self ):
        return "[{},w={},h={}]".format( self.point, self.width, self.height )
        
p = Point( 3.5, 5.0 )
r = Rectangle( p, 4.0, 2.0 )
print( r )

p.x = 1.0
p.y = 1.0
print( r )
```

Veu que canviant `p`, el `Rectangle` `r` també es canvia. El punt que conté és en realitat una *referència* al punt que es va passar al mètode `__init__()`. Igual que les llistes, els diccionaris i els conjunts, tots els objectes que són instàncies de classes que definiu, es "passen per referència" a funcions i mètodes. Per tant, en passar el `Punt` `p`, el `Rectangle` `r` es crea amb una *relació* amb `p`. D'aquesta manera es poden representar les relacions entre objectes.

No sempre vols això. De fet, és poc probable que vulgueu que un objecte "Rectangle" tingui una relació amb el punt que s'indica com a cantonada superior esquerra. Com pots resoldre això? Podeu resoldre-ho creant una còpia de l'objecte. Podeu fer-ho mitjançant el mòdul `copy`. Com s'ha comentat abans, la funció `copy()` del mòdul `copy` crea una còpia poc profunda; si voleu una còpia profunda, heu d'utilitzar la funció `deepcopy()`. Per a `Point`s això no és necessari, ja que no hi ha diferència entre les còpies superficials i profundes de les instàncies d'aquesta classe.

```python
from copy import copy

class Point:
    def __init__( self, x=0.0, y=0.0 ):
        self.x = x
        self.y = y
    def __repr__( self ):
        return "({}, {})".format( self.x, self.y )
        
class Rectangle:
    def __init__( self, point, width, height ):
        self.point = copy( point )
        self.width = width
        self.height = height
    def __repr__( self ):
        return "[{},w={},h={}]".format( self.point, self.width, self.height )
        
p = Point( 3.5, 5.0 )
r = Rectangle( p, 4.0, 2.0 )
print( r )

p.x = 1.0
p.y = 1.0
print( r )
```

(Opcional) Finalment, creeu un mètode que obtingui un segon objecte "Rectangle" com a paràmetre i retorni l'àrea superposada dels dos rectangles com un nou objecte "Rectangle" (l'últim és molt més difícil que els altres).

```python
# More complex rectangles.
from copy import copy

class Point:
    def __init__( self, x=0.0, y=0.0 ):
        self.x = x
        self.y = y
    def __repr__( self ):
        return "({}, {})".format( self.x, self.y )
        
class Rectangle:
    def __init__( self, point, width, height ):
        self.point = copy( point )
        self.width = width
        self.height = height
    def __repr__( self ):
        return "[{},w={},h={}]".format( self.point, self.width, self.height )
    def area(self):
        # width * height
        return abs(self.width) * abs(self.height)

    def perimeter(self):
        # 2 * width + 2* height
        return 2 * abs(self.width) + 2 * abs(self.height)

    def opposite_corner(self):
        # Point(x_ini + widht , y_ini - height )
        new_x = self.point.x + self.width
        new_y = self.point.y - self.height
        return Point( new_x , new_y)
        
```

```python
class Student:
    def __init__(self, name, last_name, birth_date, admin_num):
        self.name = name
        self.last_name = last_name
        self.birth_date = birth_date
        self.admin_num = admin_num
        self.courses = []

    def enroll(self, course):
        course.enroll_student( self )
        if course not in self.courses:
            self.courses.append(course)  
      
        print( f"{self.name} {self.last_name} se ha inscrito en {course.name}[{course.id}]")

    def __repr__(self):
        return f"{self.last_name}, {self.name} ({self.admin_num})"  # f-string

class Course:
    def __init__(self, name, id):
        self.name=name
        self.id = id
        self.students = []

    def enroll_student(self, student):
        if student in self.students:
            print(f'Student { student.name } is already enrolled')
        else:
            self.students.append(student)
            
    def __repr__(self):
        return f"{self.name} ({self.id})"
```
