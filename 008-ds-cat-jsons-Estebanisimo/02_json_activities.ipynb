#### load_tweets
Crea una funció anomenada "load_tweets".
La funció hauria de tenir només un paràmetre, una cadena que és la ruta relativa d'un fitxer en format "json lines".
La funció hauria d'obrir aquest fitxer i analitzar cada línia com a JSON, retornant una llista de diccionaris.

```python
import json 
def load_tweets(path):
    file = open(path)
    lines = file.readlines()
    result = []
    
    for line in lines:
        content = json.loads(line)
        result.append(content)
    file.close()
    return result

path = "tweets.json"
list_tweets = load_tweets(path)
list_tweets

```

#### get_language_counts

Crea una funció anomenada "get_language_counts" que prengui una llista de "tweets" (diccionaris) i retorni un diccionari on les claus són codis de llenguatge ("es", "ca", "en") i els valors són enters que representen el nombre de tweets en aquest idioma.

Per exemple: `{'es': 5, 'ca': 3}`

```python
def get_language_counts(tweets):
    languages = {}
    for tweet in tweets:
        lang = tweet["lang"]
        if lang is languages.keys():
            languages [lang] += 1
        elif lang is None:
            pass
        else:
            languages [lang] = 1
    return languages

get_language_counts(list_tweets)
```

Twitter enregistra els hashtags d'un tweet sota una clau anomenada "entities".

Explora el format del diccionari de tweets (Pista: utilitza l'intèrpret de Python interactiu) i descobreix com obtenir els hashtags, com a cadena, del diccionari "entities" (que conté moltes altres coses a més de hashtags).

```python
def get_hashtag_counts(tweets):
    hashtags = {}
    for tweet in tweets:
        entities = tweet["entities"] ["hashtags"]
        for ent in entities:
            ent = ent ["text"]
            
            if isinstance(ent, str) and ent in hashtags.keys():
                hashtags[ent] += 1
            elif isinstance (ent, str):
                hashtags[ent] = 1
    return hashtags

get_hashtag_counts(list_tweets)
```
